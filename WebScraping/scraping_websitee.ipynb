```python
final_df=pd.DataFrame()
for x in range(200,501):
    print("OK",x)
    url=f'https://www.ambitionbox.com/list-of-companies?page={x}'
    headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/137.0.0.0 Safari/537.36"
    ),
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Connection": "keep-alive",
    "Upgrade-Insecure-Requests": "1"}
 
    data=requests.get(url,headers=headers).text
    page=BeautifulSoup(data,'lxml')
    company = page.find_all('div',class_="companyCardWrapper")
    Name=[]
    rating=[]
    noOfReviews=[]
    salary=[]
    NoOfInterviews=[]
    jobs=[]
    benefits=[]
    photos=[]
    for i in company:
        try:
            rating.append(i.find('div',class_="rating_text").text.strip())
            Name.append(i.find('h2').get('title'))
            # rating_count.append(i.find('span',class_="companyCardWrapper__companyRatingCount").text.strip()[1:-1])
            # # noOfReviews.append(i.find('span',class_="companyCardWrapper__ActionCount").text.strip())
            d=i.find_all('div',class_="companyCardWrapper__tertiaryInformation")
            total=d[0].find_all('span',class_='companyCardWrapper__ActionCount')
            temp=[]
            for j in d[0]:
                temp.append(j.text)
            noOfReviews.append(temp[0].split(' ')[0])
            salary.append(temp[1].split(' ')[0])
            NoOfInterviews.append(temp[2].split(' ')[0])
            jobs.append(temp[3].split(' ')[0])
            benefits.append(temp[4].split(' ')[0])
            photos.append(temp[5].split(' ')[0])
        except:
            continue
    def converter(noOfReviews):
        noOfReviews=[
            f"{float(x[:-1]) * 100:g}k" if x.endswith("L") else x
            for x in noOfReviews
        ]
        return noOfReviews
    NoOfInterviews=converter(NoOfInterviews)
    df_data={'CompanyName':Name,'Rating':rating,'TotalReviews':noOfReviews,'SalaryOffereByComapany':salary,'TotalInterviews':NoOfInterviews,"Benefits":benefits,'PhotosOnSite':photos}
    df=pd.DataFrame(df_data)
    final_df = pd.concat([final_df, df], ignore_index=True)
# import pandas as pd

# final_df = pd.concat([final_df, df], ignore_index=True)
print("ok")
```
