```python
import pandas as pd
```

```python
data=pd.read_csv('ambitionbox.csv')
df=pd.DataFrame(data)
data2=pd.read_csv('ambitionbox2.csv')
df2=pd.DataFrame(data2)
```

```python
final_df=pd.concat([df,df2],ignore_index=True)
# final_df = pd.concat([final_df, df], ignore_index=True)
```

```python
final_df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
print(final_df["TotalReviews"].dtype)
```

```python
def converter(data):
    data=str(data)
    if data=='--':
        return None
    if data.endswith('L'):
        return float(data[:-1])*100000
    elif data.endswith('k'):
        return float(data[:-1])*1000
    else:
        return float(data)*1000
final_df['TotalReviews']=final_df['TotalReviews'].apply(converter)
# final_df['SalaryOffereByComapany']=final_df['SalaryOffereByComapany'].apply(converter)
```

```python
final_df['SalaryOffereByComapany']=final_df['SalaryOffereByComapany'].apply(converter)
```

```python
final_df['TotalInterviews']=final_df['TotalInterviews'].apply(converter)
```

```python
final_df['Benefits']=final_df['Benefits'].apply(converter)
```
